```python
#change release date Dtype to datetime
netflix_data['Release_Date'] = pd.to_datetime(netflix_data['Release_Date'], errors = 'coerce')

#Splitting Duration into Min/Season and Unit for future calculations
netflix_data[['Min/Season','Unit']] = netflix_data['Duration'].str.split(" " ,expand=True)

#Converting Min/Season into int
netflix_data["Min/Season"] = pd.to_numeric(netflix_data['Min/Season'], errors = 'coerce')
netflix_data.info()
```

### 1 - How many new TV shows and movies were added to Netflix in 2021?

```python
#2021 data isolation
netflix_2021 = netflix_data[netflix_data['Release_Date'].dt.year == 2021]
```

```python
#dividing into TV shows and Movies
tv_2021 = netflix_2021[netflix_2021['Category'] == 'TV Show']
movies_2021 = netflix_2021[netflix_2021['Category'] == 'Movie']
```

```python
Num_tv_2021 = len(tv_2021)
Num_movies_2021 = len(movies_2021)
```

```python
print(f"Number of new TV shows in 2021 is {Num_tv_2021}")
print(f"Number of new Movies in 2021 is {Num_movies_2021}")
```

### 2 - What is the average duration (runtime) of movies on Netflix in 2021?

```python
#using movies_2021 to collect avg duration of movies in 2021
average_movie_duration_2021 = round(movies_2021["Min/Season"].mean())
print(f"{average_movie_duration_2021} minutes")
```

### 3 - Are there any notable directors who were particularly active on Netflix in 2021?

```python
#Top 5 active Movie Directors in 2021

movie_directors = movies_2021['Director'].value_counts()

top5_movie_directors = movie_directors[movie_directors > 1].nlargest(5)

#Checks if top 5 Movie directors is empty
if top5_movie_directors.empty:
     print("All Movie directors directed one Movie only hence all Equal.")
else:
    print(f"Top 5 Movie Directors \n \n{top5_movie_directors} ")

print(f"\n\n")


#Top 5 active TV Directors in 2021
tv_show_directors = tv_2021['Director'].value_counts()

top5_tv_show_directors = tv_show_directors[tv_show_directors > 1].nlargest(5)

#Checks if top 5 Tv Show directors is empty
if top5_tv_show_directors.empty:
    print("All TV show directors directed one TV show only hence all Equal.")
else:
    print(f"Top 5 TV Show Directors \n \n{top5_tv_show_directors} ")
```
